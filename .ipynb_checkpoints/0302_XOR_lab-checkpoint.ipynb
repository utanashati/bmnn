```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
```

```python
seed = 42
np.random.seed(seed)
```

```python
bivdist = lambda mean, std, size: np.array([[i, j] for i, j in zip(np.random.normal(mean[0], std[0], size), \
                                   np.random.normal(mean[1], std[1], size))])

def generate_XOR(std, n_samples):
    pos = np.concatenate((bivdist([0, 1], [std, std], n_samples), bivdist([1, 0], [std, std], n_samples)), axis=0)
    neg = np.concatenate((bivdist([0, 0], [std, std], n_samples), bivdist([1, 1], [std, std], n_samples)), axis=0)
    return pos, neg
```

```python
std = 0.10
n_samples = 5000
pos, neg = generate_XOR(std, n_samples)
```

```python
fig, ax = plt.subplots()
ax.plot(pos[:, 0], pos[:, 1], 'gx')
ax.plot(neg[:, 0], neg[:, 1], 'ro')
ax.axis('equal')
plt.show()
```

```python
X = np.concatenate((pos, neg), axis=0)
y = np.concatenate((np.ones(pos.shape[0]), np.zeros(neg.shape[0])))
```

```python
np.random.shuffle(X)
np.random.shuffle(y)
```

```python
logs_all = []
```

```python
def models(model_id, X, y, method, epochs):
    if model_id == 1:
        model = Sequential()
        model.add(Dense(16, input_dim=2, activation='relu'))
        model.add(Dense(1, activation='sigmoid'))

        model.compile(loss='mean_squared_error',
                      optimizer='adam',
                      metrics=['binary_accuracy'])
        init_ws = model.get_weights()
#    if x == 2:   

    logs = []
    if method == 'simple':
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=seed)
        log = model.fit(X_train, y_train, epochs=epochs, verbose=0, validation_set=(X_test, y_test))
        log['params'] = [std, n_samples]
        log['model_id'] = model_id
        logs.append[log]
    if method == 'cross':
        kf = KFold(n_splits=3)
        for train_index, test_index in kf.split(X):
            log = {}
            X_train, X_test = X[train_index], X[test_index]
            y_train, y_test = y[train_index], y[test_index]
            model.set_weights(init_ws)
            model_log = model.fit(X_train, y_train, epochs=epochs, verbose=0, validation_data=(X_test, y_test))
            log['params'] = [std, n_samples]
            log['model_id'] = model_id
            log['model_log'] = model_log
            logs.append(log)
    
    return model, logs
```

```python
%time model, logs = models(1, X, y, 'cross', 500)
```

```python
logs_all.append(logs)
```

```python
colours = ['r', 'g', 'b', 'y', 'm']
```

```python
model.save('model_1')
```

```python
for ind, log in enumerate(logs_all[1]):
    plt.plot(log['model_log'].history['loss'], color=colours[ind], alpha=1)
    plt.plot(log['model_log'].history['val_loss'], color=colours[ind], linestyle='--', alpha=0.5)
plt.show()
```

```python
for ind, log in enumerate(logs_all[1]):
    plt.plot(log['model_log'].history['binary_accuracy'], color=colours[ind], alpha=1)
    plt.plot(log['model_log'].history['val_binary_accuracy'], color=colours[ind], linestyle='--', alpha=0.5)
plt.show()
```
