```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rec_tanh(init):
    arr = []
    val = init
    arr = [init]
    for i in range(1000):
        val = np.tanh(val)
        arr.append(val)
    return arr
```

```python
for i in np.arange(11)*0.1-0.5:
    plt.plot(rec_tanh(i))
```
